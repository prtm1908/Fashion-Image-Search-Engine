```python
from tensorflow.keras.models import Model
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
import cv2
import numpy as np
import os
import pickle
from tqdm import tqdm
from numpy.linalg import norm
```

```python
model=ResNet50(weights='imagenet',input_shape=(224,224,3))
model.summary()
```

```python
model_new=Model(model.input,model.layers[-2].output)
```

```python
def preprocess_img(img,model):
    img=cv2.imread(img)
    img=cv2.resize(img,(224,224))
    img=np.array(img)
    img=np.expand_dims(img,axis=0)
    img=preprocess_input(img)
    img=model.predict(img).flatten()
    img=img/norm(img)

    return img
```

```python
filename=[]
feature_list=[]

for file in os.listdir('dataset'):
    filename.append(os.path.join('Dataset',file))
```

```python
for file in filename:
    feature_list.append(preprocess_img(file,model_new))
```

```python
pickle.dump(feature_list,(open('featurevector.pkl','wb')))
pickle.dump(filename,(open('filenames.pkl','wb')))
```
